### Absolute und relative Häufigkeit der Buchstaben im Chiffrat-Text

```python
import string
import numpy as np
#create data structure to collect
character_counts = np.zeros((26,), dtype=int)
# Read file
file = open('chiffrat2.txt', 'r')

#Count frequency of characters
while 1:
    #Read one character of file
    c = file.read(1)
    #Check if c exists
    if not c:
        break
    try:
        i = string.ascii_uppercase.index(c)
        character_counts[i] += 1
    except:
        #not alpha then ignore
        pass
relative_counts = character_counts / np.sum(character_counts)

#Showcase work
for i,j in zip(string.ascii_uppercase, character_counts):
    print(i, j)

for i,j in zip(string.ascii_uppercase, relative_counts):
    print(i, j)
```

Koinzidenzindex

```python
import string
from collections import Counter
# Read file
file = open('chiffrat2.txt', 'r')

def text():
    text = []
    while 1:
        #Read one character of file
        c = file.read(1)
        #Check if c exists
        if not c:
            break
        text.append(c)
    return text

def list(text, size):
    #get emptyl lists for how much y_i we need
    groups = [[] for x in range(size)]

    #with open('chiffrat2.txt') as f:
    for i in range(len(text)):
        group = i % size
        groups[group].append(text[i])
    return groups

def f(str):
    c = dict(Counter(str))
    return c

def oben(y_j):
    f = Counter(y_j)
    sum_y_j = 0
    for key in f:
        sum_y_j += f[key] * (f[key] - 1)
    #print(sum_y_j)
    return sum_y_j

def koinzidenzindex(oben, zeile):
    unten = len(zeile) * (len(zeile)-1)
    index = oben/unten
    return index


size = int(input("What is the size of th key?", ))
grop = list(text(), size)
print("Index für die Schlüssellänge: ", size)
j = 0
for i in grop:
    c = f(i)
    #print(c)
    print("y_", j)
    j = j + 1
    print(koinzidenzindex(oben(c), i))

#8 Zeichen sind richtig
```
