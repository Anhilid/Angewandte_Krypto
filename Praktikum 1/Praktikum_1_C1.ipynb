### Absolute und relative Häufigkeit der Buchstaben in dem Chiffrat-Text

```python
import string
import numpy as np
#create data structure to collect
character_counts = np.zeros((26,), dtype=int)
# Read file
file = open('chiffrat.txt', 'r')

#Count frequency of characters
while 1:
    #Read one character of file
    c = file.read(1)
    #Check if c exists
    if not c:
        break
    try:
        i = string.ascii_uppercase.index(c)
        character_counts[i] += 1
    except:
        #not alpha then ignore
        pass
relative_counts = character_counts / np.sum(character_counts)

#Showcase work
for i,j in zip(string.ascii_uppercase, character_counts):
    print(i, j)

for i,j in zip(string.ascii_uppercase, relative_counts):
    print(i, j)
```

### Entschlüsselung des Textes

```python
pi_1 = {
        'X' : 'E',
        'R' : 'N',
        'Z' : 'I',
        'H' : 'R',
        'G' : 'S',
        'C' : 'A',
        'W' : 'T',
        'F' : 'U',
        'D' : 'H',
        'L' : 'L',
        'Q' : 'D',
        'S' : 'O',
        'U' : 'C',
        'A' : 'M',
        'M' : 'G',
        'N' : 'F',
        'V' : 'K',
        'E' : 'B',
        'O' : 'P',
        'B' : 'W',
        'P' : 'V',
        'T' : 'Z',
        'J' : 'Y',
        'Y' : 'Q',
        'K' : 'X'}

plaintext = ""
# Read file
file = open('chiffrat.txt', 'r')
# Count character occurences
while 1:
    c = file.read(1) # read one character
    if not c:
        break
    try:
        plaintext += pi_1[c]
    except:
        if c in string.ascii_uppercase:
            plaintext += "?"
        else:
            plaintext += c
file.close()

print(plaintext)
```
