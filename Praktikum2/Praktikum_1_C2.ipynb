# Decodieren von Chiffrat-Text

```python
## Read in files
# Read in chiffrat
crypt = open('chiffratbin.sec', 'rb')
c = crypt.read()

# Read in data stream
key = open('stream.dat', 'rb')
k = key.read()

##Methode to decode

#Iterate through otp data stream
for counter, i in enumerate(k):
    #get substring of the otp data stream 
    clip_otp = k[counter:(counter + len(c))]
    #xor the two strings
    xored = bytes(a ^ b for a, b in zip(clip_otp,c))

    #try making it an ascii string
    try:
        ascii_string = xored.decode('utf-8')
        print (ascii_string)
        break
    except:
        #if not ascii does not work out, reset xored
        xored = None
```
